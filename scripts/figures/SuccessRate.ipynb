```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style(
    'whitegrid', 
    {
        'grid.color': '.95',
        'axes.spines.right': False,
        'axes.spines.top': False,
    }
)

params = {
    'axes.labelsize': 9,
    'axes.titlesize':9 ,
    'font.size': 9, 
    'legend.fontsize': 6,
    'xtick.labelsize': 9, 
    'ytick.labelsize': 9,
}

plt.rcParams.update(params)
```

```python
thresholding = pd.read_csv('../../results/thresholding.csv', nrows = 4, index_col=0, decimal=',')
thresholding = thresholding.drop('Unnamed: 6', axis=1)
```

```python
steps = [8, 16, 32, 64, 128]
hydramp = thresholding.loc['HydrAMP Second phase'].tolist()
das = thresholding.loc['Das et al.']
joker = thresholding.loc['Joker']
```

```python
hydramp
```

```python
fig, axes = plt.subplots(
    ncols=1, 
    nrows=1,
    figsize=(3.75, 2.25),
    dpi=300,
)

plt.axvline(x = 32, color = 'grey', label = 'High activity', ls='--', lw=0.5,)

plt.plot(steps, hydramp, label='HydrAMP', color='#B80018', lw=1,)
plt.plot(steps, das, label='CLaSS', color='#EAAC8B', lw=1,)
plt.plot(steps, joker, label='Joker', color='#0F5257', lw=1,)

axes.set_ylabel('AMP success rate')
axes.set_xlabel(r'MIC [$\mu$g / ml]')

plt.yticks(np.arange(0, 1.04, 0.15))
plt.xticks(steps, steps)

plt.legend(bbox_to_anchor=[1, 1])
plt.show()

fig.savefig("../../../hydramp_revision/figures/thresholding_paper.svg", bbox_inches="tight")
fig.savefig("../../../hydramp_revision/figures/thresholding_paper.pdf", bbox_inches="tight")
```
